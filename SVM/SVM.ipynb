```python
"""
简化版smo
smoSimple
    dataMatIn : 每行代表一个样本
    classLabels : 类标
    C : 常数，改变这个可得到不同结果
    toler : 常数
    maxIter ：最大迭代次数
"""
s = time.clock()
b, alphas = svmMLiA.smoSimple(dataMatIn=dataArr, classLabels=labelArr, C=0.6, toler=0.001, maxIter=40)
dual = time.clock()-s
```
